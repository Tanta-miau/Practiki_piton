# Задание 1

Задача: Создать чат бота для получения информации об исследованиях космоса

Описание: Создайте комплексное приложение командной строки, которое будет использоваться в качестве панели управления исследованиями космоса. Данное приложение будет обращаться к https://api.nasa.gov/ для предоставления пользователям набора информации о космосе, включая:

- Астрономическая картинка дня (APOD): Отображение APOD с пояснениями к нему.
- Фотографии с марсохода: позволяет пользователям выбирать и фильтровать фотографии с марсохода по дате и типу камеры.
- Объекты, сближающиеся с Землей (ОСЗ): Поиск и отображение объектов, сближающихся с Землей, на определенную дату, включая их размеры и потенциальную опасность.
- Данные о космической погоде: Отображают последние данные о космической погоде, включая солнечные вспышки и геомагнитные бури.
Приложение должно позволять пользователям ориентироваться в этих функциях, корректно обрабатывать ошибки и обеспечивать удобство работы.

Требования:
- Пользовательский ввод: Приложение должно предложить пользователю ввести данные, чтобы выбрать, какую функцию он хочет изучить.
- Проверка данных: Убедитесь, что пользовательские данные (например, даты) проверены.
- Обработка ошибок: Корректно обрабатывайте ошибки API и неверные ответы.
- Представление данных: Представляйте данные в четкой и организованной форме.
- Опция выхода: позволяет пользователям выходить из приложения в любое время.

```python
import requests
import datetime as dt

#тут работает только астрономическая картинка и космическая погода

time=dt.datetime.utcnow()#сегодняшняя дата(для вспышек)
segodnya=time.strftime('%Y-%m-%d')
vchera=time - dt.timedelta(hours=24)#вчерашняя дата
vchera=vchera.strftime('%Y-%m-%d')

print('введите запрос. Астрономическая картинка дня или Данные о космической погоде')
zapros=input()
#Астрономическая картинка дня
if zapros=='Астрономическая картинка дня':
  url='https://api.nasa.gov/planetary/apod?api_key=DEMO_KEY'
  response = requests.get(url)
  ss=response.text.split('explanation')
  ss=(((ss)[1])[3:])
  ss1=ss.split(',"hdurl')  #для вывода текста
  foto=ss.split('url')   #для вывода фото
  foto=(foto[2])[2:]
  foto=foto[:-2]
  print('информация:',(ss1[0])[:-1])
  print(f'фото: {foto}')

#elif zapros=='Фотографии с марсохода':
elif zapros=='fot':
  url='https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?'


#elif zapros=='Объекты, сближающиеся с Землей' or zapros=='ОСЗ':
elif zapros=='os':
  url='https://ssd-api.jpl.nasa.gov/nhats.api'





#Данные о космической погоде
if zapros=='Данные о космической погоде':
  #геомагнитные бури
  url='https://api.nasa.gov/DONKI/GST?startDate=yyyy-MM-dd&endDate=yyyy-MM-dd&api_key=DEMO_KEY'
  response = requests.get(url)
  del_link=response.text.split('link')
  first_gms=((del_link[1])[:-3])[3:]
  second_gms=((del_link[3])[:-3])[3:]
  print('геомагнитные бури. начало: ',first_gms)
  print('геомагнитные бури. конец: ',second_gms)


  #вспышки на солнце
  print('информация о вспышках на солнце')
  url='https://api.nasa.gov/DONKI/FLR?startDate=yyyy-MM-dd&endDate=yyyy-MM-dd&api_key=DEMO_KEY'
  response = requests.get(url)
  dell=response.text.split('submissionTime')
  dlina=len(dell)
  for odna_vsp in range(1,dlina):
    stroka=(dell[odna_vsp]).split('linkedEvents')
    stroka=stroka[0]

    if str(vchera) in stroka:
      stroka=((stroka)[2:])[:-3]
      stroka=stroka.split('T')
      stroka=stroka[1].split('Z')
      print('вчера:',stroka[0],((stroka[1])[24:]))
    elif str(segodnya) in stroka:
      stroka=((stroka)[2:])[:-3]
      stroka=stroka.split('T')
      stroka=stroka[1].split('Z')
      print('сегодня:',stroka[0],((stroka[1])[24:]))

```
